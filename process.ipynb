# load Data of Products

```python
import numpy as np
dfproduct=pd.read_csv('/content/drive/MyDrive/Oportunidades/Products_completo.csv')
dfproduct.replace('Preferente','SI', regex=True, inplace=True)
dfproduct.replace('Si','SI', regex=True, inplace=True)
dfproduct.replace('Sí','SI', regex=True, inplace=True)
dfproduct.rename(columns={"2a14": "12a14"},inplace=True)
dfproduct.drop(columns=['SAP','UPC Unidad de venta','Gr./ Ml','Unidad Gr.'],inplace=True)
dfproduct
```

```python
laliprod=dfproduct['Descripción Lean'].unique().tolist()
```

## Searching parameters for $[P]$ in $[S]$
### Normalize Formula for Rank:
#### $$\tilde{x}_i = \frac{x_i - x_{min}}{x_{max} - x_{min}}$$

```python
lidf=[]
for x in laliprod:
  dfprod=dfproduct[dfproduct['Descripción Lean']==x]
  row = 0                                                                                                                                                                                     
  valmac ='SI'
  list_macro = [c for c in dfprod.columns if dfprod[c].iloc[row] == valmac]                                                                                                                                   
  filtered_macro = dfprod[list_macro]                                                                                                                                                                  
  lasmacro=filtered_macro.columns.tolist()                                                                                                                                                                                    
  value = 1                                                                                                                                                                                
  list_comp = [c for c in dfprod.columns if dfprod[c].iloc[row] == value]                                                                                                                                   
  filtered_df = dfprod[list_comp]                                                                                                                                                                  
  lascol=filtered_df.columns.tolist()
  edad=[]
  nse=[]
  for r in lascol:
    if r in [ '0a2', '3a5','6a11', '12a14', '15a17', '18a24', '25a40', '41a50', '51a60', '61a70','mayor70']:
      edad.append(r)
    else:
      nse.append(r)
  dfsumedad=dfjoined[edad]
  dfsumnse=dfjoined[nse]

  dfsumedad['Ponderar_edad']= dfsumedad.iloc[:,:].astype(float).sum(axis=1)
  dfsumnse['Ponderar_nse']= dfsumnse.iloc[:,:].astype(float).sum(axis=1)
          
  dfjoined['Ponderar_edad']=dfsumedad['Ponderar_edad']
  dfjoined['Ponderar_nse']=dfsumnse['Ponderar_nse']
  df_final=dfjoined[['key','poblacion_total', 'Macroformato',
        'SumaNse', 'Ponderar_edad', 'Ponderar_nse']]
  def comprueba(fila):
    if fila.Macroformato in lasmacro and (fila.SumaNse>0 and float(fila.poblacion_total)>0) : 
      pond_nse=(fila.Ponderar_nse*50)/float(fila.SumaNse)
      pond_pob=(fila.Ponderar_edad*50)/float(fila.poblacion_total)
      prom_pond=(pond_nse+pond_pob)/2
    else:
      prom_pond=0
    return prom_pond

  df_final['prom_pond']=df_final.apply(comprueba,axis=1)
  df_final['Normalizacion']=1+(((df_final['prom_pond']-df_final['prom_pond'].min())*(5-1))/(df_final['prom_pond'].max()-df_final['prom_pond'].min()))
  df_final['Normalizacion'] = df_final['Normalizacion'].replace(np.nan, 1)
  df_final['rank']=df_final.Normalizacion.astype(str).str[0:1]
  df_final['Descripción Lean']=dfprod['Descripción Lean'].iloc[0]
  df_final['Descripción SAP']=dfprod['Descripción SAP'].iloc[0]
  df_final['Categoría']=dfprod['Categoría'].iloc[0]
  df_final['Marca']=dfprod['Marca'].iloc[0]
  df_final['UPC x Fracción']=dfprod['UPC x Fracción'].iloc[0]
  df_final.columns=['Tienda', 'poblacion_total', 'Macroformato', 'Total_Nse', 'Suma_interes_edad',
        'Suma_interes_Nse', 'prom_pond', 'Normalizacion', 'rank',
        'Descripción Lean', 'Descripción SAP', 'Categoría', 'Marca','UPC x Fracción']

  lidf.append(df_final)
  

```
